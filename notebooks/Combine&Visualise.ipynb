```python
df1 = pd.read_csv("data/raw/rajasthan 2-5-2023-2026.csv")
df2 = pd.read_csv("data/raw/rajasthan 6-8-2020 - 1-5-2023.csv")
df3 = pd.read_csv("data/raw/rajasthan 15-9-2017--7-6-2020.csv")
df4 = pd.read_csv("data/raw/rajasthan 20-12-2014 - 14-9-2020 .csv")
df5 = pd.read_csv("data/raw/rajasthan 1-1-2014 - 19-12-2014.csv")
```

```python
combine_df = pd.concat([df1,df2, df3, df4, df5] , ignore_index = True)
```

```python
selected_columns = ['datetime', 'tempmax', 'temp', 'humidity', 'windspeed','winddir',
                    'precip', 'dew', 'cloudcover', 'solarradiation']

df = combine_df[selected_columns]
```

```python
df.head()
```

```python
cols = ['tempmax', 'temp', 'dew']

for col in cols:
    df[col] = (df[col] - 32) / 1.8
    df[col] = df[col].round(2)
    
```

```python
df['windspeed'] = df['windspeed'] * 1.60934
```

```python
import numpy as np

def calculate_relative_humidity(T, Dp):
    beta = 17.625
    lam = 243.04
    
    RH = 100 * (
        np.exp((beta * Dp) / (lam + Dp)) /
        np.exp((beta * T) / (lam + T))
    )
    return RH
```

```python
df['calculated_RH'] = calculate_relative_humidity(
    df['temp'],
    df['dew']
)
```

```python
df['calculated_RH'].round(2)
df['calculated_RH'] = df['calculated_RH'].clip(0, 100)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='mixed', dayfirst=True)
df['month'] = df['datetime'].dt.month
```
