```python
selected_columns = ['datetime', 'tempmax', 'temp', 'feelslikemax', 'humidity', 'windspeed','winddir',
                    'precip', 'dew', 'cloudcover', 'solarradiation']

df = combine_df[selected_columns]
```

```python
df.head()
```

```python
cols = ['tempmax', 'temp', 'feelslikemax', 'dew']

for col in cols:
    df[col] = (df[col] - 32) / 1.8
    df[col] = df[col].round(2)
    
```

```python
df['windspeed'] = df['windspeed'] * 1.60934
```

```python
import numpy as np

def calculate_relative_humidity(T, Dp):
    beta = 17.625
    lam = 243.04
    
    RH = 100 * (
        np.exp((beta * Dp) / (lam + Dp)) /
        np.exp((beta * T) / (lam + T))
    )
    return RH
```

```python
df['calculated_RH'] = calculate_relative_humidity(
    df['temp'],
    df['dew']
)
```

```python
df['calculated_RH'].round(2)
df['calculated_RH'] = df['calculated_RH'].clip(0, 100)
```

```python
df[['humidity', 'calculated_RH']].describe()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='mixed', dayfirst=True)
df['month'] = df['datetime'].dt.month
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df_hot['datetime'], df_hot['tempmax'])
plt.title("Daily Maximum Temperature – Rajasthan (Summer)")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
monthly_avg = df_hot.groupby('month')['tempmax'].mean()

monthly_avg.plot(kind='bar', title="Average Summer Temperature (Rajasthan)")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
df_hot['heatwave'] = df_hot['tempmax'] >= 41
```

```python
df_hot['year'] = df_hot['datetime'].dt.year
```

```python

heatwave_days = df_hot.groupby('year')['heatwave'].sum()

plt.figure(figsize=(10,4))
plt.scatter(heatwave_days.index, heatwave_days.values, marker='o')
plt.title("Year-wise Heatwave Days in Rajasthan")
plt.xlabel("Year")
plt.ylabel("Number of Heatwave Days")
plt.show()
```
